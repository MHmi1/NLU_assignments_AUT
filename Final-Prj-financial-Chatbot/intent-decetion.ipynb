```python
from pathlib import Path

# Function to check the format of a dataset file
def check_dataset_format(file_path):
    """
    Check each line in the file for the correct format: 'utterance_data <=> intent_label'.
    Logs errors with file name and line number.
    """
    errors = []
    lines = Path(file_path).read_text("utf-8").strip().splitlines()

    for line_number, line in enumerate(lines, start=1):
        # Check if the line contains the delimiter " <=> "
        if " <=> " not in line:
            errors.append(f"File: {file_path}, Line: {line_number}, Content: {line}")

    return errors

# Main function to check multiple files
def check_all_files(file_paths):
    """
    Check all specified files for format issues.
    """
    all_errors = []
    for file_path in file_paths:
        print(f"Checking file: {file_path}...")
        errors = check_dataset_format(file_path)
        if errors:
            all_errors.extend(errors)
        else:
            print(f"No issues found in {file_path}.")
    return all_errors

# Define file paths
train_file = "../train.txt"
validation_file = "../validation.txt"
test_file = "../test.txt"


# Check all files
file_paths = [train_file, validation_file, test_file]
errors = check_all_files(file_paths)

# Print errors
if errors:
    print("\nFormat Errors Detected:")
    for error in errors:
        print(error)
else:
    print("\nAll files are correctly formatted!")
```

```python
import json
import random
import torch
from torch.utils.data import DataLoader, TensorDataset
from transformers import XLMRobertaForSequenceClassification, XLMRobertaTokenizer, AdamW
from torch.nn import CrossEntropyLoss
import numpy as np
from tqdm import tqdm
from sklearn.metrics import f1_score, precision_recall_fscore_support, accuracy_score
from pathlib import Path
import matplotlib.pyplot as plt

# Tokenizer initialization
from transformers import AutoModel, AutoTokenizer
MODEL_NAME = "/home/mh/Documents/NLU-exe/xlm_roberta/xlm-roberta-base"  # Use XLM-Roberta for Persian
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, add_prefix_space=True)

def encode_data(sentences, intents, tokenizer, intent2id):
    """
    Encode sentences and map intents to their corresponding IDs.
    """
    encodings = tokenizer(sentences, padding=True, truncation=True, max_length=128, return_tensors='pt')
    intent_labels = [intent2id[intent] for intent in intents]
    return encodings, torch.tensor(intent_labels)


# Convert encodings to tensors for DataLoader
def convert_to_tensor(encodings, labels):
    input_ids = encodings['input_ids']
    attention_mask = encodings['attention_mask']
    labels = labels
    return TensorDataset(input_ids, attention_mask, labels)

# Training loop
def train_epoch(model, data_loader, optimizer, device, loss_fn):
    model.train()
    total_loss = 0
    correct_predictions = 0

    for data in tqdm(data_loader):
        input_ids, attention_mask, labels = [t.to(device) for t in data]

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()

        logits = outputs.logits
        _, preds = torch.max(logits, dim=1)
        correct_predictions += torch.sum(preds == labels).item()

        loss.backward()
        optimizer.step()

    return correct_predictions / len(data_loader.dataset), total_loss / len(data_loader)

# Evaluation loop
def eval_model(model, data_loader, device):
    model.eval()
    all_preds = []
    all_labels = []
    total_loss = 0

    with torch.no_grad():
        for data in data_loader:
            input_ids, attention_mask, labels = [t.to(device) for t in data]

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()

            logits = outputs.logits
            _, preds = torch.max(logits, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculate both micro and macro metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision_micro, recall_micro, f1_micro, _ = precision_recall_fscore_support(all_labels, all_preds, average='micro')
    precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(all_labels, all_preds, average='macro')

    return {
        'accuracy': accuracy,
        'precision_micro': precision_micro,
        'recall_micro': recall_micro,
        'f1_micro': f1_micro,
        'precision_macro': precision_macro,
        'recall_macro': recall_macro,
        'f1_macro': f1_macro,
        'loss': total_loss / len(data_loader)
    }

# Function to predict intent for a single example
def predict_example(text, tokenizer, model, id2intent):
    """
    Predict the intent of a given text.
    """
    model.eval()
    encoding = tokenizer(text, padding=True, truncation=True, max_length=128, return_tensors='pt')
    input_ids = encoding['input_ids'].to(device)
    attention_mask = encoding['attention_mask'].to(device)

    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)

    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=1).item()
    predicted_intent = id2intent[predicted_class]
    return predicted_intent

# Function to preprocess data (convert slot-labeled data to plain text)
def preprocess_data(lines):
    """
    Convert dataset lines to plain text by removing slot labels and intents.
    """
    sentences = []
    intents = []

    for line_number, line in enumerate(lines, start=1):
        # Check if the line contains the delimiter " <=> "
        if " <=> " not in line:
            print(f"Skipping malformed line at Line {line_number}: {line}")
            continue

        # Split into utterance and intent
        utterance_data, intent_label = line.split(" <=> ")

        # Remove slot labels (e.g., ":o", ":b-fname") from the utterance
        plain_text = " ".join([word.split(":")[0] for word in utterance_data.split()])
        sentences.append(plain_text)
        intents.append(intent_label.strip())

    return sentences, intents

# Load and process data
if __name__ == "__main__":
    # Define file paths
    train_file = "../train.txt"
    validation_file = "../validation.txt"
    test_file = "../test.txt"

    # Load data
    def load_data(file_path):
        lines = Path(file_path).read_text("utf-8").strip().splitlines()
        return preprocess_data(lines)

    # Preprocess and load data
    train_sentences, train_intents = load_data(train_file)
    val_sentences, val_intents = load_data(validation_file)
    test_sentences, test_intents = load_data(test_file)

    # Create intent-to-id mapping
    intent2id = {intent: idx for idx, intent in enumerate(sorted(set(train_intents + val_intents)))}

    # Encode data
    train_encodings, train_labels = encode_data(train_sentences, train_intents, tokenizer, intent2id)
    val_encodings, val_labels = encode_data(val_sentences, val_intents, tokenizer, intent2id)

    # Convert to TensorDataset
    train_dataset = convert_to_tensor(train_encodings, train_labels)
    val_dataset = convert_to_tensor(val_encodings, val_labels)

    # DataLoader for batch processing
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)

    # Model initialization
    model = XLMRobertaForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=len(intent2id))
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)

    # Define optimizer
    optimizer = AdamW(model.parameters(), lr=2e-5)

    # Training loop with early stopping
    patience = 2
    best_val_f1 = float('-inf')
    early_stop_counter = 0

    epochs = 9
    train_accuracies = []
    train_losses = []
    val_accuracies = []
    val_losses = []

    for epoch in range(epochs):
        print(f"Epoch {epoch + 1}/{epochs}")
        train_acc, train_loss = train_epoch(model, train_loader, optimizer, device, CrossEntropyLoss())
        print(f"Train loss: {train_loss}, accuracy: {train_acc}")

        val_metrics = eval_model(model, val_loader, device)
        print(f"Validation Metrics:")
        print(f"  Accuracy: {val_metrics['accuracy']}")
        print(f"  Micro - Precision: {val_metrics['precision_micro']}, Recall: {val_metrics['recall_micro']}, F1: {val_metrics['f1_micro']}")
        print(f"  Macro - Precision: {val_metrics['precision_macro']}, Recall: {val_metrics['recall_macro']}, F1: {val_metrics['f1_macro']}")

        # Save metrics for plotting
        train_accuracies.append(train_acc)
        train_losses.append(train_loss)
        val_accuracies.append(val_metrics['accuracy'])
        val_losses.append(val_metrics['loss'])

        # Early stopping check
        if val_metrics['f1_macro'] > best_val_f1:
            best_val_f1 = val_metrics['f1_macro']
            early_stop_counter = 0
            model.save_pretrained('./best_intent_model')  # Save the best model
            tokenizer.save_pretrained('./best_intent_model')
        else:
            early_stop_counter += 1
            if early_stop_counter >= patience:
                print("Early stopping triggered.")
                break

    # Plot training and validation accuracy/loss
    epochs_range = range(1, len(train_accuracies) + 1)
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, train_accuracies, label='Train Accuracy')
    plt.plot(epochs_range, val_accuracies, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, train_losses, label='Train Loss')
    plt.plot(epochs_range, val_losses, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()

    # Apply the trained model on test.txt
    print("\nProcessing test.txt for intent detection...")
    id2intent = {v: k for k, v in intent2id.items()}
    for sentence in test_sentences:
        predicted_intent = predict_example(sentence, tokenizer, model, id2intent)
        print(f"Text: {sentence}\nPredicted Intent: {predicted_intent}\n{'-' * 50}")
```

```python
import random
from pathlib import Path
import json
import torch
from torch.utils.data import DataLoader, TensorDataset
from transformers import XLMRobertaForSequenceClassification, XLMRobertaTokenizer, AdamW
from torch.nn import CrossEntropyLoss
import numpy as np
from tqdm import tqdm
from sklearn.metrics import f1_score, precision_recall_fscore_support, accuracy_score
import matplotlib.pyplot as plt

# Tokenizer initialization
from transformers import AutoModel, AutoTokenizer
MODEL_NAME = "/home/mh/Documents/NLU-exe/xlm_roberta/xlm-roberta-base"  # Use XLM-Roberta for Persian
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, add_prefix_space=True)

# Function to append new data to the existing dataset files
def append_new_data(new_data_file, train_file, validation_file, test_file):
    """
    Append new data from `new_data_file` to `train_file`, `validation_file`, and `test_file`
    while maintaining the proportions: 70% train, 20% validation, 10% test.
    """
    # Read the new data
    with open(new_data_file, "r", encoding="utf-8") as f:
        new_lines = f.readlines()

    # Shuffle the new data
    random.shuffle(new_lines)

    # Split the new data
    total_records = len(new_lines)
    train_split = int(0.7 * total_records)
    validation_split = int(0.9 * total_records)  # 70% train + 20% validation

    train_data = new_lines[:train_split]
    validation_data = new_lines[train_split:validation_split]
    test_data = new_lines[validation_split:]

    # Function to append data to a file
    def append_to_file(data, file_name):
        with open(file_name, "a", encoding="utf-8") as f:  # Open in append mode
            f.writelines(data)

    # Append the data to the respective files
    append_to_file(train_data, train_file)
    append_to_file(validation_data, validation_file)
    append_to_file(test_data, test_file)

    print(f"New data successfully appended to {train_file}, {validation_file}, and {test_file}.")

# Function to load and preprocess data
def load_data(file_path):
    """
    Load and preprocess data from a file.
    Each line is expected to be in the format: 'utterance_data <=> intent_label'.
    """
    lines = Path(file_path).read_text("utf-8").strip().splitlines()
    return preprocess_data(lines)

# Function to preprocess data (convert slot-labeled data to plain text)
def preprocess_data(lines):
    """
    Convert dataset lines to plain text by removing slot labels and intents.
    """
    sentences = []
    intents = []

    for line_number, line in enumerate(lines, start=1):
        # Check if the line contains exactly one " <=> " delimiter
        if line.count(" <=> ") != 1:
            print(f"Skipping malformed line at Line {line_number}: {line}")
            continue

        # Split into utterance and intent
        try:
            utterance_data, intent_label = line.split(" <=> ")
        except ValueError:
            print(f"Skipping line due to splitting error at Line {line_number}: {line}")
            continue

        # Remove slot labels (e.g., ":o", ":b-fname") from the utterance
        plain_text = " ".join([word.split(":")[0] for word in utterance_data.split()])
        sentences.append(plain_text)
        intents.append(intent_label.strip())

    return sentences, intents

def convert_to_tensor(encodings, labels):
    """
    Convert tokenized encodings and labels into a TensorDataset.
    
    Args:
        encodings (dict): Tokenized inputs (e.g., 'input_ids', 'attention_mask').
        labels (torch.Tensor): Tensor of intent label IDs.
    
    Returns:
        TensorDataset: A PyTorch dataset containing input_ids, attention_mask, and labels.
    """
    input_ids = encodings['input_ids']
    attention_mask = encodings['attention_mask']
    return TensorDataset(input_ids, attention_mask, labels)

# Function to shuffle the training data
def shuffle_data(sentences, intents):
    """
    Shuffle the sentences and intents together to ensure randomization.
    """
    combined = list(zip(sentences, intents))
    random.shuffle(combined)
    shuffled_sentences, shuffled_intents = zip(*combined)
    return list(shuffled_sentences), list(shuffled_intents)

def encode_data(sentences, intents, tokenizer, intent2id):
    """
    Encode sentences and map intents to their corresponding IDs.
    
    Args:
        sentences (list of str): The input sentences to be tokenized.
        intents (list of str): The intent labels corresponding to each sentence.
        tokenizer: The tokenizer (e.g., XLM-Roberta tokenizer) used to tokenize the sentences.
        intent2id (dict): A dictionary mapping intent labels to their unique IDs.
    
    Returns:
        encodings (dict): Tokenized inputs with keys 'input_ids', 'attention_mask', etc.
        intent_labels (torch.Tensor): Tensor of intent label IDs.
    """
    # Tokenize the sentences
    encodings = tokenizer(
        sentences,
        padding=True,           # Pad sentences to the same length
        truncation=True,        # Truncate sentences longer than max_length
        max_length=128,         # Maximum sequence length
        return_tensors='pt'     # Return PyTorch tensors
    )
    
    # Map intent labels to their corresponding IDs
    intent_labels = [intent2id[intent] for intent in intents]
    
    # Convert intent labels to a PyTorch tensor
    return encodings, torch.tensor(intent_labels)

def train_epoch(model, data_loader, optimizer, device, loss_fn):
    """
    Perform one epoch of training.

    Args:
        model: The model to train.
        data_loader: DataLoader for the training data.
        optimizer: Optimizer for updating the model weights.
        device: Device to run the training on (e.g., 'cuda' or 'cpu').
        loss_fn: Loss function to compute the loss.

    Returns:
        Tuple containing:
            - Training accuracy for the epoch.
            - Average training loss for the epoch.
    """
    model.train()
    total_loss = 0
    correct_predictions = 0

    for data in tqdm(data_loader):
        input_ids, attention_mask, labels = [t.to(device) for t in data]

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()

        logits = outputs.logits
        _, preds = torch.max(logits, dim=1)
        correct_predictions += torch.sum(preds == labels).item()

        loss.backward()
        optimizer.step()

    return correct_predictions / len(data_loader.dataset), total_loss / len(data_loader)

# Evaluation loop
def eval_model(model, data_loader, device):
    model.eval()
    all_preds = []
    all_labels = []
    total_loss = 0

    with torch.no_grad():
        for data in data_loader:
            input_ids, attention_mask, labels = [t.to(device) for t in data]

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()

            logits = outputs.logits
            _, preds = torch.max(logits, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculate both micro and macro metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision_micro, recall_micro, f1_micro, _ = precision_recall_fscore_support(all_labels, all_preds, average='micro')
    precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(all_labels, all_preds, average='macro')

    return {
        'accuracy': accuracy,
        'precision_micro': precision_micro,
        'recall_micro': recall_micro,
        'f1_micro': f1_micro,
        'precision_macro': precision_macro,
        'recall_macro': recall_macro,
        'f1_macro': f1_macro,
        'loss': total_loss / len(data_loader)
    }

def predict_example(text, tokenizer, model, id2intent):
    """
    Predict the intent of a given text.

    Args:
        text (str): The input text to predict the intent for.
        tokenizer: The tokenizer used to encode the text.
        model: The trained model for intent classification.
        id2intent (dict): A dictionary mapping intent IDs to their corresponding intent labels.

    Returns:
        str: The predicted intent label.
    """
    model.eval()
    encoding = tokenizer(text, padding=True, truncation=True, max_length=128, return_tensors='pt')
    input_ids = encoding['input_ids'].to(device)
    attention_mask = encoding['attention_mask'].to(device)

    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)

    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=1).item()
    predicted_intent = id2intent[predicted_class]
    return predicted_intent

# Load and process data
if __name__ == "__main__":
    train_file = "../train.txt"
    validation_file = "../validation.txt"
    test_file = "../test.txt"
    new_data_file = "../new_data.txt"  # New data to append

    # Append new data to the existing dataset files
    append_new_data(new_data_file, train_file, validation_file, test_file)

    # Load and preprocess the extended datasets
    train_sentences, train_intents = load_data(train_file)
    val_sentences, val_intents = load_data(validation_file)
    test_sentences, test_intents = load_data(test_file)

    # Shuffle the training data to avoid bias
    train_sentences, train_intents = shuffle_data(train_sentences, train_intents)

    # Create intent-to-id mapping
    intent2id = {intent: idx for idx, intent in enumerate(sorted(set(train_intents + val_intents)))}

    # Encode data
    train_encodings, train_labels = encode_data(train_sentences, train_intents, tokenizer, intent2id)
    val_encodings, val_labels = encode_data(val_sentences, val_intents, tokenizer, intent2id)

    # Convert to TensorDataset
    train_dataset = convert_to_tensor(train_encodings, train_labels)
    val_dataset = convert_to_tensor(val_encodings, val_labels)

    # DataLoader for batch processing
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)

    # Model initialization
    model = XLMRobertaForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=len(intent2id))
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)

    # Define optimizer
    optimizer = AdamW(model.parameters(), lr=2e-5)

    # Training loop with early stopping
    patience = 2
    best_val_f1 = float('-inf')
    early_stop_counter = 0

    epochs = 9
    train_accuracies = []
    train_losses = []
    val_accuracies = []
    val_losses = []

    for epoch in range(epochs):
        print(f"Epoch {epoch + 1}/{epochs}")
        train_acc, train_loss = train_epoch(model, train_loader, optimizer, device, CrossEntropyLoss())
        print(f"Train loss: {train_loss}, accuracy: {train_acc}")

        val_metrics = eval_model(model, val_loader, device)
        print(f"Validation Metrics:")
        print(f"  Accuracy: {val_metrics['accuracy']}")
        print(f"  Micro - Precision: {val_metrics['precision_micro']}, Recall: {val_metrics['recall_micro']}, F1: {val_metrics['f1_micro']}")
        print(f"  Macro - Precision: {val_metrics['precision_macro']}, Recall: {val_metrics['recall_macro']}, F1: {val_metrics['f1_macro']}")

        # Save metrics for plotting
        train_accuracies.append(train_acc)
        train_losses.append(train_loss)
        val_accuracies.append(val_metrics['accuracy'])
        val_losses.append(val_metrics['loss'])

        # Early stopping check
        if val_metrics['f1_macro'] > best_val_f1:
            best_val_f1 = val_metrics['f1_macro']
            early_stop_counter = 0
            model.save_pretrained('./best_intent_model_bot_challenge')  # Save the best model
            tokenizer.save_pretrained('./best_intent_model_bot_challenge')
        else:
            early_stop_counter += 1
            if early_stop_counter >= patience:
                print("Early stopping triggered.")
                break

    # Plot training and validation accuracy/loss
    epochs_range = range(1, len(train_accuracies) + 1)
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, train_accuracies, label='Train Accuracy')
    plt.plot(epochs_range, val_accuracies, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, train_losses, label='Train Loss')
    plt.plot(epochs_range, val_losses, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()

    # Apply the trained model on test.txt
    print("\nProcessing test.txt for intent detection...")
    id2intent = {v: k for k, v in intent2id.items()}
    for sentence in test_sentences:
        predicted_intent = predict_example(sentence, tokenizer, model, id2intent)
        print(f"Text: {sentence}\nPredicted Intent: {predicted_intent}\n{'-' * 50}")
```
